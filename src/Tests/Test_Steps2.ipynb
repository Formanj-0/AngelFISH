```python
import os
import sys
cwd = os.getcwd()
source_path = os.path.join(cwd, '..')
print(source_path)
sys.path.append(source_path)

from source import Data, Parameters, Pycro, SimpleCellposeSegmentaion, SpotDetection_Steps
```

```python
params = Parameters()
data = Data('DEFAULT_NAME')

params.num_chunks_to_run = 10
```

```python
step = Pycro()

params.load_in_mask = True
params.local_dataset_location =  [os.path.join(os.getcwd(), '..', r'dataBases\JF001_0min_5uM_12')]
params.initial_data_location = None
params.independent_params = {}

output = step.main(**params.__dict__)
data.append(output)
print(data)
```

```python
import numpy as np
import dask.array as da

pp = data.images.shape[0]
tt = data.images.shape[1]
cc = data.images.shape[2]
zz = data.images.shape[3]
yy = data.images.shape[4]
xx = data.images.shape[5]


step = SimpleCellposeSegmentaion()
params.nucChannel = 2
params.cytoChannel = None
if data.nuc_mask is None:
    data.nuc_mask = da.zeros([pp, tt, zz, yy, xx], chunks=(1, 1, zz, yy, xx))

count = 0
for p in range(pp):
    for t in range(tt):
        if count >= params.num_chunks_to_run:
            break
        kwargs = {**params.__dict__, **data.__dict__}
        kwargs['image'] = data.images[p, t]
        kwargs['fov'] = p
        kwargs['timepoint'] = t
        output = step.main(**kwargs)
        data.nuc_mask[p, t, 0, :, :] = output['nuc_mask']
        count += 1


```

```python
from joblib import Parallel, delayed

def process_chunk(p, t, params):
    output = SimpleCellposeSegmentaion().main(**params)
    return p, t, output['nuc_mask']

results = Parallel(n_jobs=-1)(
    delayed(process_chunk)(
        p, t, {**params.__dict__, **data.__dict__, 'image': data.images[p, t], 'fov': p, 'timepoint': t}, 
    )
    for p in range(data.images.shape[0])
    for t in range(data.images.shape[1])
    if p * data.images.shape[1] + t < params.num_chunks_to_run
)

for p, t, nuc_mask in results:
    data.nuc_mask[p, t, 0, :, :] = nuc_mask

```
